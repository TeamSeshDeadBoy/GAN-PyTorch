# Визуализируем фото, сгенерированные моделями

## 0. Imports

```python
import PIL
import matplotlib.pyplot as plt
from pathlib import Path
```

## 1. Определяем список путей фотографии

```python
model = "DCGAN" # Тут выбрать название модели ["GAN" / "ProGAN" / "DCGAN"]
```

```python
paths = list(Path(f"../images/{model}").glob('*.png'))
len(paths)
```

```python
for i in range(20, len(paths)+1, 20):
    plt.figure(figsize=(10, 8))
    plt.title(f"Generated by {model}   [{i/20}/{len(paths)/20}]")
    for ind, path in enumerate(paths[i-20:i]):
        image = PIL.Image.open(path).convert("RGBA")
        plt.subplot(4, 5, ind+1)
        plt.axis(False)
        plt.imshow(image)
```
